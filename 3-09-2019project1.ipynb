```python
import pandas as pd 
df=pd.read_csv("clustering.csv")
df.head()
```

```python
final=df.drop("Loan_ID",axis=1)
final.head()
```

```python
final.Gender=final.Gender.fillna("Female")
final.head()
```

```python
final.Self_Employed=final.Self_Employed.fillna("No")
final.head()
```

```python

final.Loan_Amount_Term=final.Loan_Amount_Term.median()
final.Credit_History=final.Credit_History.median()
final.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label1=LabelEncoder()
label2=LabelEncoder()
label3=LabelEncoder()
label4=LabelEncoder()
label5=LabelEncoder()
label6=LabelEncoder()
label1.fit_transform(final.Gender)
label2.fit_transform(final.Married)
label3.fit_transform(final.Education)
label4.fit_transform(final.Self_Employed)
label5.fit_transform(final.Property_Area)
label6.fit_transform(final.Loan_Status)

final.Gender=label1.fit_transform(final.Gender)
final.Married=label2.fit_transform(final.Married)
final.Education=label3.fit_transform(final.Education)
final.Self_Employed=label4.fit_transform(final.Self_Employed)
final.Property_Area=label5.fit_transform(final.Property_Area)
final.Loan_Status=label6.fit_transform(final.Loan_Status)
```

```python
final.head()
```

```python
x=final.drop(["Loan_Status","Dependents"],axis=1)
x.head()
y=final.Loan_Status
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.svm import SVC
model=SVC(C=70)
model.fit(x_train,y_train)
```

```python
yn=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,yn)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(facecolor='yellow',figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel("predicted")
plt.ylabel("truth")
plt.show()
```

```python
model.score(x_test,y_test)
```
