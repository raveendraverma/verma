```python
import matplotlib.pyplot as plt
x=[2,3,6]
y=[5,6,7]
plt.fill(x,y,label="first",linewidth="10",linestyle="dotted",color="pink")
plt.bar([3,4,5],[5,6,7],label="second",linewidth="2")
plt.xlabel("x_coordinate")
plt.ylabel("y_coordinate")
plt.show()
```
